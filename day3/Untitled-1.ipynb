```python
valid = []

for i in range(len(data)):
    if data[i:i+4] == "mul(":
        if "," in data[i+4:i+8] and ")" in data[i+5:i+12]:
            end_index = data.find(")", i, i+12)
            valid.append(data[i+4:end_index])
    if i == len(data) - 1:
        break

valid
```

```python
results = 0
for pairs in valid:
    sublist = pairs.split(",")
    num1, num2 = sublist
    if num1.isalnum and num2.isalnum:
        results += int(num1)*int(num2)

results
```
